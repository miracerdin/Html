## Soru 1 -->  1'den 10000 bine kadar Amicable (Dostane) sayıları yazınız. 
## Amicable numbers --> a'nın tam bölenleri  = b , b'nin tam bölenleri = a




```python
amacable = []
for i in range(10000,0,-1):
    top = 0
    for j in range(1,10000):
        if i % j == 0:
            top += j
    top -= i
    top2 = 0
    for y in range(1, top+1):
        if top % y == 0:
            top2 += y
    top2 -= y

    if top2 == i and top == y and i != y and [i,y] not in amacable:
        amacable.append(sorted([i,y]))
print(amacable)
```

```python
def division(a):
    b = 0
    for i in range(1,a):
        if a % i == 0:
            b += i
    return b
liste = []
x = 0
while x <10000:
    if division(division(x)) == x and division(x) != x:
        liste.append(x)
    x += 1
print(liste)
```
